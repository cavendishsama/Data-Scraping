```python
import re

# titles[0]

# Function to extract the university name
def extract_university_name(data):
    # Regular expression to match everything before the first |
    pattern = r"^(.*?)\s*\|"
    university_names = [re.match(pattern, item).group(1) for item in data if re.match(pattern, item)]
    return university_names

# Test the function
extracted_names = extract_university_name(titles)
# print(extracted_names)
# print(extracted_names[0])
```
